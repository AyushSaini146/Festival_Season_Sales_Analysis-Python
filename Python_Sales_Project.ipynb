# Exploratory Data Analysis(EDA)

# 1] Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

#for count on columns
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount spend
Sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=Sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men



# 2] Age Group

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Age Group vs Total Amount Spent
Sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=Sales_age)
```

Above graphs shows that most of the buyers are of age group between 26-35 yrs female

# 3] State

```python
#Top 10 States according to total orders made
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ab=sns.barplot(data=sales_state,x='State',y='Orders')

# Add count labels on top of each bar
for p in ab.patches:
    ab.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

```python
#Total sales From Top 10 States
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ab=sns.barplot(data=sales_state,x='State',y='Amount')

# Add count labels on top of each bar
for p in ab.patches:
    ab.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

From above graphs we can see that most of the orders and sales are from Uttar Pradesh,mAharashtra and Karnataka respectively.

# 4] Marital Status

```python
#Marital Status of the customers
ax=sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount Spent vs Marital status
Marital_Sales = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
ab=sns.barplot(data = Marital_Sales, x = 'Marital_Status',y= 'Amount', hue='Gender')

# Add count labels on top of each bar
for p in ab.patches:
    ab.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# 5] Occupation of customers

```python
#Ocuupation of customers
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(15,5)})

#Count Labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Occupation Vs Purchasing Power
Marital_Sales = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
ab=sns.barplot(data = Marital_Sales, x = 'Occupation',y= 'Amount')

# Add count labels on top of each bar
for p in ab.patches:
    ab.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sectors

# 6] Product Category

```python
#Total orders
sns.set(rc={'figure.figsize':(22,7)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Product Category vs Total Sales(amount) 
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ab=sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')

# Add count labels on top of each bar
for p in ab.patches:
    ab.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
#Top 10 Most Selling Products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
